```python
from transformers import pipeline, set_seed

model_id = "BSC-LT/salamandra-2b"
```

```python
import pandas as pd
from transformers import pipeline
import re  # Importar el módulo re para expresiones regulares

# Cargar los archivos CSV
#df_plats = pd.read_csv(plats)  # Reemplaza con tu ruta
df_jap = pd.read_csv(jap)  # Reemplaza con tu ruta

# Cargar el modelo de Hugging Face
modelo = pipeline("text-generation", model=model_id)

# Función para generar un contexto utilizando el modelo
def generar_contexto(prefLabel, writtenRep):
    prompt = f"Genera un context descriptiu per al terme '{writtenRep}' que és un tipus de plat de '{prefLabel}'. Proporciona un context únic i rellevant perquè la paraula no sigui ambigua en aquest context."

    # Generar el contexto
    respuesta = modelo(prompt, num_return_sequences=1)
    contexto = respuesta[0]['generated_text']

    return contexto.strip()

# Función para procesar writtenRep
def procesar_writtenRep(writtenRep):
    # Sustituir guiones bajos por espacios
    writtenRep = writtenRep.replace('_', ' ')
    # Eliminar cualquier texto que esté dentro de corchetes
    writtenRep = re.sub(r'\[.*?\]', '', writtenRep)  # Elimina todo lo que está entre corchetes
    return writtenRep.strip()  # Retornar el resultado sin espacios extra

# Guardar progresivamente los contextos a medida que se generan
def guardar_contextos_progresivos(df, archivo_salida):
    # Abrir el archivo en modo 'append' si ya existe, o crear uno nuevo si no
    with open(archivo_salida, 'a') as f:
        # Iterar por cada fila del DataFrame
        for index, row in df.iterrows():
            # Tomar solo la primera representación escrita (writtenRep) y procesarla
            primera_writtenRep = row['writtenRep'].split(',')[0]  # Suponiendo que writtenRep puede tener múltiples valores separados por comas
            primera_writtenRep = procesar_writtenRep(primera_writtenRep)  # Procesar la representación

            # Generar el contexto si no existe
            if 'contexto' not in row or pd.isna(row['contexto']):
                contexto = generar_contexto(row['prefLabel'], primera_writtenRep)

                # Guardar solo el contexto en el archivo CSV
                f.write(f"{contexto}\n")

                print(f"Contexto generado para '{primera_writtenRep}' y guardado.")

# Crear o actualizar archivos CSV de contextos progresivamente
#guardar_contextos_progresivos(df_plats, 'contextos_plats.csv')
guardar_contextos_progresivos(df_jap, "contextos_jap_1.csv")

print("Contextos generados y guardados con éxito.")

```
